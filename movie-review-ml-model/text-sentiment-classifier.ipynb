```python
movie_data = load_files('/Users/pramanathan/query-resource-predictor/review_polarity/txt_sentoken')
```

```python
X, y = movie_data.data, movie_data.target
```

```python
y
```

```python
documents = []

from nltk.stem import WordNetLemmatizer

stemmer = WordNetLemmatizer()

for sen in range(0, len(X)):
    # Remove all the special characters
    document = re.sub(r'\W', ' ', str(X[sen]))
    
    # remove all single characters
    document = re.sub(r'\s+[a-zA-Z]\s+', ' ', document)
    
    # Remove single characters from the start
    document = re.sub(r'\^[a-zA-Z]\s+', ' ', document) 
    
    # Substituting multiple spaces with single space
    document = re.sub(r'\s+', ' ', document, flags=re.I)
    
    # Removing prefixed 'b'
    document = re.sub(r'^b\s+', '', document)
    
    # Converting to Lowercase
    document = document.lower()
    
    # Lemmatization
    document = document.split()

    document = [stemmer.lemmatize(word) for word in document]
    document = ' '.join(document)
    
    documents.append(document)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(max_features=1500, min_df=5, max_df=0.7, stop_words=stopwords.words('english'))
```

```python
stopwords.words('english')
```

```python
X = vectorizer.fit_transform(documents).toarray()
```

```python
X[1]
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidfconverter = TfidfTransformer()
X = tfidfconverter.fit_transform(X).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=1000, random_state=0)
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
 def clean_text(str_data):
        # Remove all the special characters
        str_data = re.sub(r'\W', ' ', str_data)

        # remove all single characters
        str_data = re.sub(r'\s+[a-zA-Z]\s+', ' ', str_data)

        # Remove single characters from the start
        str_data = re.sub(r'\^[a-zA-Z]\s+', ' ', str_data) 

        # Substituting multiple spaces with single space
        str_data = re.sub(r'\s+', ' ', str_data, flags=re.I)

        # Removing prefixed 'b'
        str_data = re.sub(r'^b\s+', '', str_data)

        # Converting to Lowercase
        str_data = str_data.lower()

        # Lemmatization
        str_data = str_data.split()

        str_data = [stemmer.lemmatize(word) for word in str_data]
        str_data = ' '.join(str_data)
        return str_data
```
